```python
# This simulates what a real webhook would send to your system
fake_webhook_payload = {
    "event_type": "new_customer_signup",
    "customer": {
        "email": "jane@example.com",
        "name": "Jane Doe",
        "plan": "premium"
    },
    "timestamp": "2026-04-14T14:30:00Z"
}

print("Fake webhook created:")
print(fake_webhook_payload)
```

```python
def process_webhook(payload):
    """Extract key information from webhook payload"""
    event = payload.get("event_type")
    customer_email = payload.get("customer", {}).get("email")
    customer_name = payload.get("customer", {}).get("name")
    
    alert_message = f"ALERT: {event} from {customer_name} ({customer_email})"
    return alert_message

result = process_webhook(fake_webhook_payload)
print(result)
```

```python
def process_webhook_with_rules(payload):
    event = payload.get("event_type")
    customer_email = payload.get("customer", {}).get("email")
    customer_name = payload.get("customer", {}).get("name")
    plan = payload.get("customer", {}).get("plan", "basic")
    
    if event == "new_customer_signup":
        if plan == "premium":
            priority = "HIGH"
        else:
            priority = "NORMAL"
        alert = f"[{priority}] New signup: {customer_name} - {plan} plan"
    
    elif event == "cancellation":
        alert = f"URGENT: {customer_name} cancelled"
    
    else:
        alert = f"INFO: {event} from {customer_email}"
    
    return alert

print(process_webhook_with_rules(fake_webhook_payload))
```

```python
import smtplib
from email.message import EmailMessage


GMAIL_USER = "lehlohonolomtshizana@gmail.com"  
GMAIL_APP_PASSWORD = "my_password"   

msg = EmailMessage()
msg["Subject"] = "Automation Alert: Webhook Received"
msg["From"] = GMAIL_USER
msg["To"] = GMAIL_USER  
msg.set_content(f"Alert: {process_webhook_with_rules(fake_webhook_payload)}\nCustomer: {fake_webhook_payload['customer']['name']}")


try:
    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
        server.login(GMAIL_USER, GMAIL_APP_PASSWORD)
        server.send_message(msg)
        print("✅ Email sent via Gmail!")
except Exception as e:
    print(f"❌ Failed: {e}")
```
