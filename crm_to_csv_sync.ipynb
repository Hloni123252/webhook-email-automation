```python
import requests
import pandas as pd
from datetime import datetime

# Connect to HubSpot public API
url = "https://api.hubapi.com/crm/v3/objects/contacts?limit=10"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    print(f"✅ Connected! Found {len(data.get('results', []))} contacts")
    
    # Extract contact data
    contacts = []
    for item in data.get('results', []):
        props = item.get('properties', {})
        contacts.append({
            'firstname': props.get('firstname', ''),
            'lastname': props.get('lastname', ''),
            'email': props.get('email', ''),
            'company': props.get('company', '')
        })
    
    # Save to CSV
    df = pd.DataFrame(contacts)
    filename = f"crm_export_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
    df.to_csv(filename, index=False)
    print(f"✅ Saved {len(contacts)} contacts to {filename}")
    print("\nPreview:")
    print(df.head())
else:
    print(f"❌ API Error: {response.status_code}")
```
